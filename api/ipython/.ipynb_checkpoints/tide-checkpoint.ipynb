```python
DATAID = "F-A0021-001"
CWBKEY = "CWB-574A0062-1D41-409B-9046-DDA4D1EDD4A8"
URL = "http://opendata.cwb.gov.tw/opendataapi?dataid={}&authorizationkey={}".format(DATAID, CWBKEY)
```

```python
import requests
r = requests.get(URL)
print URL
```

```python
import xmltodict, json
o = xmltodict.parse(r.text)
qq = json.dumps(o, ensure_ascii=False, indent=2)
#print qq
```

```python
data = {}
```

```python
o.keys()
```

```python
for each in o['cwbopendata']['dataset']['location']:
    data.update({each['locationName']: []})
    for time in each['time']:
        time['validTime']['startTime']
        data[each['locationName']].append(time)
```

```python
for each in data[u"屏東縣枋寮鄉"][0]['weatherElement'][2]['time']:
    for x in each['weatherElement']:
        print json.dumps(x,ensure_ascii=False)
```

```python
from collections import OrderedDict
_data = OrderedDict()
for each in data[u"屏東縣枋寮鄉"][0]['weatherElement'][2]['time']:
    tide = each['weatherElement'][0]['value']
    time = each['weatherElement'][1]['value']
    height = each['weatherElement'][3]['elementValue']['value']
    _data.update({time:[tide, height]})
```
