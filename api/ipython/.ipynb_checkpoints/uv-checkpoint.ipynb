```python
import requests
url = "http://opendata.epa.gov.tw/webapi/api/rest/datastore/355000000I-000004?sort=PublishTime&offset=0&limit=1000"
r = requests.get(url)
```

```python
import json
j = json.loads(r.text)
print json.dumps(j, ensure_ascii=False, indent=2)
```

```python
type(j)
```

```python
data = {}
for each in j['result']['records']:
    data.update({each['County']:each['UVI']})
```

```python
print json.dumps(data,ensure_ascii=False, indent=2)
```
